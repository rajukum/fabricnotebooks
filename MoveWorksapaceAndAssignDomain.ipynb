```python
import sempy_labs as labs
import sempy_labs.admin as admin
import sempy.fabric as fabric

# Define the target and source capacities
source_capacity_id = "F6BAC257-3178-4F7B-93C2-AA1A8FBDE367"  # Replace with the actual source capacity ID In Caps
target_capacity_id = "c89438cd-cfa3-4ff9-be46-6f2f2e673cbf"  # Fabric capacity ID for the IT domain in small

# Define the domain and subfolder names
domain_name = "IT"
subfolder_name = "IT-Migration"

# Define the list of workspace names to move
workspace_names_to_move = [
    "ASMigration"
    #"Admin monitoring",
    #"Azure Cost Management App"
    # Add more workspace names as needed
]
vfilter = "capacityId eq '" + source_capacity_id + "'"
workspaces_in_capacity = fabric.list_workspaces(vfilter)

#display(workspaces_in_capacity)

existing_capacities = admin.list_capacities()

#display(existing_capacities)

if target_capacity_id not in existing_capacities['Capacity Id'].values:
    raise ValueError(f"The specified target capacity '{target_capacity_id}' does not exist.")
else:
    target_capacity_names = existing_capacities[existing_capacities['Capacity Id']==target_capacity_id]
    #display(target_capacity_names)

target_capacity_name = target_capacity_names['Capacity Name'].values    
print(target_capacity_name[0])

for workspace in workspace_names_to_move:
    # Step 1: Assign workspace to the target capacity
    labs.assign_workspace_to_capacity(
        capacity_name =target_capacity_name[0],
        workspace =workspace
    )
    print(f"Workspace {workspace} assigned to capacity {target_capacity_name[0]}.")

    # Step 2: Organize the workspace within the IT domain and the specified subfolder
    try:
        # Move the workspace to the IT domain
        admin.assign_domain_workspaces(
            domain_name =subfolder_name,
            workspace_names =workspace
        )
        print(f"Workspace '{workspace}' assigned to domain '{subfolder_name}'.")

        # Move the workspace to the specific subfolder within the domain
    except AttributeError as e:
        print(f"Error organizing workspace '{workspace}': {e}")
```
